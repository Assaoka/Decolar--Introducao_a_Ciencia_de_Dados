# üß† Tomando Decis√µes no C√≥digo

Imagine um programa como um caminho. √Äs vezes, esse caminho √© uma linha reta, mas na maioria das vezes, ele tem bifurca√ß√µes. O programa precisa decidir qual caminho seguir com base nas informa√ß√µes que recebe.

# Estruturas de Sele√ß√£o: `if`, `elif` e `else`

Pense nessas estruturas como um plano com v√°rias etapas:
- **`if` (Se):** √â o "Plano A". O c√≥digo verifica se uma condi√ß√£o √© verdadeira. Se for, ele executa o bloco de c√≥digo do `if` e ignora todo o resto.
- **`elif` (Se n√£o, se):** √â o "Plano B", "Plano C", etc. Se o `if` (ou um `elif` anterior) for falso, o programa testa a condi√ß√£o do `elif`. Se for verdadeira, executa o bloco dele e para por a√≠.
- **`else` (Se n√£o):** √â a op√ß√£o "√∫ltimo caso". Se nenhuma das condi√ß√µes anteriores (do `if` e dos `elifs`) for verdadeira, o bloco do `else` ser√° executado.

Veja como esse fluxo funciona em Python:

```python
condicao_1 = True
condicao_2 = True

if condicao_1:
    # Bloco de c√≥digo a ser executado se a condicao_1 for verdadeira
    print("Executei o Plano A!")

elif condicao_2:
    # Bloco executado se a condicao_1 for FALSA e a condicao_2 for VERDADEIRA
    print("O Plano A falhou, ent√£o executei o Plano B!")

else:
    # Bloco executado se TODAS as condi√ß√µes anteriores forem falsas
    print("Nenhum plano deu certo, ent√£o executei o √∫ltimo caso.")
```

**‚ö†Ô∏è Ponto de Aten√ß√£o:** Observe os dois pontos (`:`) no final de cada linha de `if`, `elif` e `else`. Eles s√£o obrigat√≥rios! Igualmente importante √© a **indenta√ß√£o** (o espa√ßo ou "tab" antes do `print`). Em Python, √© a indenta√ß√£o que define quais comandos pertencem a cada bloco.

# Operadores Relacionais (Os S√≠mbolos de Compara√ß√£o)

Ok, mas o que s√£o essas "condi√ß√µes"? Geralmente, s√£o perguntas que resultam em True (Verdadeiro / Sim) ou False (Falso / N√£o).

| Operador | Descri√ß√£o                                                             | Exemplo                                                |
| -------- | --------------------------------------------------------------------- | ------------------------------------------------------ |
| `==`     | Verifica se dois valores s√£o iguais.                                  | `1 == 1` ‚Üí `True`<br><br>`0 == 1` ‚Üí `False`            |
| `!=`     | Verifica se dois valores s√£o diferentes.                              | `0 != 1` ‚Üí `True`                                      |
| `>`      | Verifica se o valor da esquerda √© maior que o valor da direita.       | `2 > 1` ‚Üí `True`<br><br>**Aten√ß√£o:** `2 > 2` ‚Üí `False` |
| `<`      | Verifica se o valor da esquerda √© menor que o valor da direita.       | `1 < 2` ‚Üí `True`                                       |
| `>=`     | Verifica se o valor da esquerda √© maior ou igual ao valor da direita. | `2 >= 2` ‚Üí `True`                                      |
| `<=`     | Verifica se o valor da esquerda √© menor ou igual ao valor da direita. | `1 <= 2` ‚Üí `True`                                      |

# Operadores L√≥gicos: `and`, `or` e `not`

Viu como o c√≥digo anterior come√ßou a ficar um pouco confuso, com um `if` dentro do outro? Parece um "Haduken" de c√≥digo!

```python
if condicao1:
	if condicao2:
		if condicao3:
			if condicao4:
				print("Haduken")
```

Em vez de criar um `if` dentro de outro `if` (o que pode deixar o c√≥digo confuso), podemos combinar as compara√ß√µes usando operadores l√≥gicos.

| Operador | Descri√ß√£o                                                                        |
| -------- | -------------------------------------------------------------------------------- |
| `and`    | Retorna verdadeiro se ambas as express√µes condicionais forem verdadeiras.        |
| `or`     | Retorna verdadeiro se pelo menos uma das express√µes condicionais for verdadeira. |
| `not`    | Retorna o inverso da express√£o condicional. (V ‚û° F, F ‚û° V).                      |
